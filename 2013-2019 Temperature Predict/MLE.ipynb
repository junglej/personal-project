```python
import pandas as pd
precip = pd.read_csv("../data/changping.txt", index_col=0, na_values='NA', delim_whitespace=True)
#display the first 5 rows
precip.head(5)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#do some data exploration with the histogram
_ = precip.hist(sharex=True, sharey=True, grid=False)
plt.tight_layout()
```

```python
#data cleansing
precip.fillna(value={'RH': precip.RH.mean()}, inplace=True)
precip
```

```python
#the means of data in precip
precip_mean = precip.mean()
precip_mean
```

```python
#the variances of data in precip
precip_var = precip.var()
precip_var
```

```python
#give the two formulas to get two numbers, which can be used in gamma distribution
alpha_mom = (precip_mean**2)/(precip_var**2)
beta_mom = (precip_var**2)/precip_mean
alpha_mom, beta_mom
```

```python
#use the gamma.pdf function in scipy.stats.distributions to plot the distributions implied by the calculated alphas and betas.
from scipy.stats.distributions import gamma

precip.wind.hist(normed=True, bins=20)
plt.plot(np.linspace(0, 10), gamma.pdf(np.linspace(0, 10), alpha_mom[0], beta_mom[0]))
```

```python
#create a grid of plots for the distribution, using the gamma distribution
axs = precip.hist(normed=True, figsize=(12, 8), sharex=True, sharey=True, bins=15, grid=True)

for ax in axs.ravel():

    # Get month
    m = ax.get_title()

    # Plot fitted distribution
    x = np.linspace(*ax.get_xlim())
    ax.plot(x, gamma.pdf(x, alpha_mom[m], beta_mom[m]))

    # Annotate with parameter estimates
    label = 'alpha = {0:.2f}\nbeta = {1:.2f}'.format(alpha_mom[m], beta_mom[m])
    ax.annotate(label, xy=(10, 0.2))

plt.tight_layout()
```

```python
#only keep the selected data
data = data[data.wind>3]
RH = data.RH
```

```python
#plotting the raw survival data and the histogram
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

ax1.plot(sorted(RH)[::-1], 'o')
ax1.set_xlabel('number of days')
ax1.set_ylabel('RH')

ax2.hist(RH, bins=15)
ax2.set_xlabel('RH')
ax2.set_ylabel('number of days')
```

```python
#compute this parameter
smean =RH.mean()
rate = 1. / smean
rate
```

```python
smax = RH.max()
days = np.linspace(0., smax, 1000)
dt = smax / 999.
dist_exp = st.expon.pdf(days, scale=1. / rate)

#rescale the theoretical distribution to the histogram
nbins = 30
fig, ax = plt.subplots(1, 1, figsize=(6, 4))
ax.hist(RH, nbins)
ax.plot(days, dist_exp * len(RH) * smax / nbins,
        '-r', lw=3)
ax.set_xlabel("RH")
ax.set_ylabel("Number of days")
```
